```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# sns.barplot(x='sex', y='total_bill', data=tips)
# by default estimator is an average/mean function apllied on numeric column, we may change 
# it with our own custom function e.g.
import numpy as np
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

```python
# Count plot
sns.countplot(x='sex', data=tips)
```

```python
# Box Plot
sns.boxplot(x='day', y='total_bill', data=tips)
# sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
# sns.violinplot(x='day', y='total_bill', data=tips)
# sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
```

```python
# strip plot
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
# swarm plot
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
# Factor plot : General form of all the plots explained above
sns.factorplot(x='day', y='total_bill', data=tips)
```
