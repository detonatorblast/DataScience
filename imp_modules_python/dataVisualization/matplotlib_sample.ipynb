```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,5, 11)
```

```python
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional Approach for plotting
plt.plot(x, y)
plt.xlabel("X Label")
plt.ylabel("Y label")
plt.title("sample plot using functional approach")
type(plt)
```

```python
# multiple plots using functional approach
plt1 = plt.subplot(1,2,1)
plt1.plot(x,y)
plt1.set_title("First Plot")

plt2 = plt.subplot(1,2,2)
plt2.plot(y, x, 'r')
plt2.set_title("Second Plot")
type(plt2)
```

# Object Oriented approach to plot using matplotlib

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Sample OO plot")

axes.plot(x,y)
```

```python
# we may draw multiple plots using OO approach also
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y)

axes2 = fig.add_axes([0.2, 0.4, 0.3, 0.4])
axes2.plot(x, y)
```

```python
# we may use subplots to draw multiple graphs
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# for a in axes:
#     a.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title("First plot")

axes[1].set_title("second plot")
axes[1].plot(y,x)
```

# Figure Size and DPI

```python
# we may pass size of canvas for plotting
fig = plt.figure(figsize=(8,2))
axes = fig.add_axes([0, 0, 1, 1])

axes.plot(x,y)
```

```python
# we may pass the figsize to subplots
fig, axes = plt.subplots(figsize=(8,2), nrows=2, ncols=1)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig.savefig("samplePlot.png", dpi=100)
```

# legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_title("sample plot with legends")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")

ax.plot(x, x**2, label='sqaured')
ax.plot(x, x**3, label='cubed')

ax.legend()
```
